## Condicionais e estruturas iterativas

---

Estruturas condicionais são utilizadas para verificar o estado de condições (Verdadeiro ou Falso) e executar um comando a depender do estado:

````python
a=2
b=2
if a==1 and b==1:
  print('a e b são iguais a 1')
elif a==2 and b==2:
  print('a e b são iguiais a 2')
else:
  print('a='+str(a)+' and b='+str(b))
````

Estruturas iterativas são utilizadas para a repetição de um determinado bloco de instrução:

````python
a=[5,3,[1,2],'Hi']
for obj in a:
	print(obj)
````

````python
for i in range(len(a)):
	print(i)
````

````python
for i,obj in enumerate(a):
  print('{} - {}'.format(i,obj))
````

````python
[obj for obj in a]
````

````python
[[i,obj] for i,obj in enumerate(a)]
````

````python
# Importar bibliotecas a serem utilizadas
import numpy as np
from matplotlib import pyplot as plt
````

````python
#Desenhando um gráfico.
t=np.linspace(0,8,200)
w=np.pi
y=np.sin(w*t)
plt.plot(t,y,'b--',linewidth=1)
plt.plot(t,np.zeros(len(t)),'k:',linewidth=1)
plt.plot(t,y,'ko',markersize=5,fillstyle='none')
plt.legend(['1°','2°','3°'])
plt.title('Graphic Name')
plt.ylabel('$sin(\omega t)$ [-]')
plt.grid()
plt.xlim([min(t),max(t)])
plt.text(1,0.75,'HERE',fontsize=16)
plt.savefig('sin.png',dpi=300)
````

````python
	#Desenhando vários gráficos em uma mesma imagem.
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.grid()
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.grid()
plt.xlabel('time (s)')
plt.ylabel('Undamped')
````

````python
#Desenhando e interpolando imagens.
A = np.random.rand(10, 10)
fig, ax = plt.subplots(1, 3, figsize=(10, 3))
list_interp=['nearest', 'bilinear', 'bicubic']
for i,interp in enumerate(list_interp):
  ax[i].imshow(A, interpolation=interp)
  ax[i].set_title(interp.capitalize())
  ax[i].grid(True)
````

### Manipulação de equações algébricas

````python
x = symbols('x')
expr = x**2 +2*x+1
Eq(expr,0)
````

````python
solveset(expr,x)
````

````python
expr1 = '(x**2 + 2*x + 1)/x'
expr1 = sympify(expr1)
expr1
````

### Limites, derivadas e integrais

````python
func, limit_value = sin(x)/x, 0
Eq(Limit(func, x, limit_value, '+'), limit(func, x, limit_value, '+'))
````

````python
Eq(Derivative(cos(x),(x,1)),diff(cos(x),x,1))
````

````python
Derivative(sin(x),(x,1)).doit().subs([(x,pi)])#.evalf()
````

````python
Eq(Integral(cos(x),(x,0,1)),integrate(cos(x),(x, 0, 1)))
````

````python
integrate(cos(x),(x,0,pi/2)).evalf()
````

````python
f,n = symbols('f',cls=Function), symbols('n')
eq_derivada = f(x).diff((x,n))
eq_derivada
````

````python
eq_derivada = eq_derivada.subs(n, 2).replace(f, lambda x: sin(x)+x**2)
eq_derivada
````

````python
eq_derivada.doit()
````

````python
m, k, c, x, t = symbols('m k c x t')
diffeq = Eq(m*f(t).diff(t,2) + c*f(t).diff(t,1) + k*f(t) , 0)
diffeq
````

````python
ddiffeq = dsolve(diffeq, f(t))
sol = ddiffeq.subs([(c,100),(m,5),(k,4000)])
sol
````

3. Plotar a equação algébrica ($s$), considerando $a$ = - 9,81 m/s | $s_0$ = 20 m e $v_0$ = 0 m/s.


Dicas

```
1) Importe a biblioteca sympy e inicialize as variáveis com "symbols";
2) Utilize a função "Eq(lado esquerdo, lado direito)"  para expressar a equação;
3) Derive com Derivative / diff;
4) Integre com Integral / integrate;
5) Substitua os valores de a, s0 e t0 na equação (s) utilizando o comando func.subs([(var, valor]);
6) Plote a função (s) com plotting.plot(func, (var, x0, x1))
```

````python
from sympy import *
# Primeiro modo
s, s0, v0, t, a = symbols('s s0 v0 t a')
eq = Eq(s , s0+v0*t+(a*t**2)/2)
eq
````

````python
Eq(Derivative(eq.rhs,t),diff(eq.rhs,t,1))
````

````python
Eq(Integral(eq.rhs,(t,0,t)), integrate(eq.rhs,(t,0,t)))
````

````python
eq1 = eq.subs([(a,-9.81),(s0,20),(v0,0)])
eq1
````

````python
plotting.plot(eq1.rhs, (t, 0, 2), line_color='k', title='Movimento em queda Livre',
          xlabel='Tempo [s]', ylabel='Altitude [m]', size=(6,4.5))
````

````python
def linear_func(x, a, b):
  y = a*x+b
  plt.scatter(x, y)
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('$y={:.1f}x+{:.1f}$'.format(a, b))
  plt.grid()
  return y
x = np.arange(10)
a, b = 10.2, 10.5
y_new=linear_func(x, a, b)
````

````python
def sin_func(t, A, w, phi):
  y = A*np.sin(w*t+phi)
  plt.plot(t, y)
  plt.xlabel('tempo')
  plt.ylabel('$y={}sin({}t+{})$'.format(A, w,phi))
  plt.grid()
  return y
````
